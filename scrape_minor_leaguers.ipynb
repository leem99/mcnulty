# Download Minor League Player Pages from BBREF

```python
import requests
import os
from bs4 import BeautifulSoup
from fake_useragent import UserAgent
import pandas as pd
import numpy as np
import re
import time
import csv
import itertools
```

Make Helper functions

```python
def random_soup(url):
    us = UserAgent()
    user_agent = {'User-Agent':us.random}
    response = requests.get(url,headers = user_agent)
    page = response.text
    soup = BeautifulSoup(page,"lxml")
    return soup
```

Connect to master list of all minor leagues

```python
# This page contains a table with all minors
url = 'https://www.baseball-reference.com/register/league.cgi'
main_soup = random_soup(url)
```

```python
league_table_html = main_soup.find('table')
league_table_rows = league_table_html.find_all('tr')
```

Loop through all table to find all league pages

```python
good_urls = []
#keep_leagues = ['AAA','AA','A%2B','A','A-','Rk']
#keep_leagues = ['A%2B']
#keep_leagues = ['A']
#keep_leagues = ['A-']
keep_leagues = ['AA']

for ix, row in enumerate(league_table_rows):
    if ix >0:
        if row.find_all('td'):
            if int(row.find_all('td')[1].string) >= 1990:
                temp_url = row.find('a')['href']
                
                if ((temp_url.split("class=",1)[1] in keep_leagues)
                    and ('MEX' not in temp_url)):
                    
                    good_urls.append(row.find('a')['href'])
```

Loop through league pages to find team pages

```python
team_season_urls = []
for gu in good_urls:
    time.sleep(np.random.uniform(0,1))
    us = UserAgent()
    user_agent = {'User-Agent':us.random}
    league_url = 'https://www.baseball-reference.com'+ gu
    league_soup = random_soup(league_url)

    league_table = league_soup.find('table')
    rows = league_table.find_all('tr')

    for ix, row in enumerate(rows):
        if ix > 0:
            temp_links = row.find_all('a')
            year = int(temp_links [0].text)
            if year >= 1990:
                for item in temp_links[1:]:
                    team_season_urls.append(item['href'])
            else:
                break
    
```

Loop through team pages to find player pages

```python
print(len(team_season_urls))
# File to keep track of finished teams
teams_urls_finished = []
batter_links = []
```

```python
with open('milb_batter_links.csv', 'a+', newline='') as csvfile:
    player_link_writer = csv.writer(csvfile, delimiter=',')
      
    for ix, tu in enumerate(team_season_urls):
        if tu not in teams_urls_finished:
            time.sleep(np.random.uniform(0,1)) 
            us = UserAgent()
            user_agent = {'User-Agent':us.random}
            team_url = 'https://www.baseball-reference.com'+ tu
            team_soup = random_soup(team_url)
            team_batting_table = team_soup.find('table')
            temp_links = team_batting_table.find_all('a',href=re.compile(r'/register/player'))
            for bl in temp_links:
                batter_links.append(bl['href'])
                player_link_writer.writerow([bl['href']])
                #print(bl)
            teams_urls_finished.append(tu)
            print(str(ix)+ ' Team Link:' + tu)
    csvfile.close()
```

Read back the file

```python
with open('milb_batter_links.csv', 'r') as f:
    reader = csv.reader(f)
    p_list = list(reader)
    f.close()

batter_links = list(set(list(itertools.chain.from_iterable(p_list))))
```

Save all of the player links in a csv file

```python
saved_flist = os.listdir('milb_player_pages')
```

```python
len(batter_links)
```

```python
## Loop through files and download
# Put in delays so that I don't overload website
for ix, bl in enumerate(batter_links):
    fname = re.sub(r'\W+', '', bl.split('=')[-1])
    if fname+'.html' not in saved_flist:
        time.sleep(1+np.random.uniform(0,1)) 
        batter_url = 'http://www.baseball-reference.com'+bl
        us = UserAgent()
        user_agent = {'User-Agent':us.random}
        response = requests.get(batter_url,headers = user_agent)
        batter_page = response.text
        
        with open("milb_player_pages/"+fname+'.html', "w") as pfile:
            pfile.write(batter_page)
            pfile.close()
            
            
        saved_flist.append(fname)
        print(ix, fname)
```
